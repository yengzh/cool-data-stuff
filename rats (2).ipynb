```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

df = pd.read_csv("CFW.csv")

freqs = [4, 8, 12, 16, 24, 32]
ages = {
    '1 mo':  ['01mo_04_khz', '01mo_08_khz', '01mo_12_khz', '01mo_16_khz', '01mo_24_khz', '01mo_32_khz'],
    '6 mo':  ['06mo_04_khz', '06mo_08_khz', '06mo_12_khz', '06mo_16_khz', '06mo_24_khz', '06mo_32_khz'],
    '10 mo': ['10mo_04_khz', '10mo_08_khz', '10mo_12_khz', '10mo_16_khz', '10mo_24_khz', '10mo_32_khz'],
}
colors = {'1 mo': '#E8A882', '6 mo': '#5BAD8F', '10 mo': '#B0C4D8'}
timepoint_cols = {
    '1 mo':  '01mo_test_date',
    '6 mo':  '06mo_test_date',
    '10 mo': '10mo_test_date',
}
age_labels = list(ages.keys())
```

```python
x_range = np.linspace(0, 110, 300)

fig, axes = plt.subplots(len(freqs), 1, figsize=(8, 14), sharex=True)

for row, (ax, freq_idx) in enumerate(zip(axes, range(len(freqs)))):
    ax.axhline(0, color='black', linewidth=1.2)
    ax.set_ylabel(f'{freqs[freq_idx]} kHz', rotation=0, labelpad=45, va='center', fontsize=11)
    ax.set_yticks([])
    ax.spines[['top', 'right', 'left']].set_visible(False)

    for age_label, col_list in ages.items():
        col = col_list[freq_idx]
        color = colors[age_label]
        for sex, sign in [('F', 1), ('M', -1)]:
            vals = pd.to_numeric(df[df['sex'] == sex][col].replace('NR', np.nan), errors='coerce').dropna()
            if len(vals) < 5:
                continue
            kde = gaussian_kde(vals, bw_method=0.3)
            density = kde(x_range)
            density = density / density.max() * 0.4 * sign
            ax.fill_between(x_range, 0, density, alpha=0.6, color=color)
            ax.plot(x_range, density, color=color, linewidth=0.8)

    if row == 0:
        ax.text(55, 0.32, 'females', fontsize=9)
        ax.text(55, -0.35, 'males', fontsize=9)

legend_patches = [plt.matplotlib.patches.Patch(color=colors[a], label=a) for a in ages]
axes[-1].set_xlabel('ABR threshold (dB)', fontsize=12)
fig.legend(handles=legend_patches, loc='upper right', frameon=False)
plt.tight_layout()
plt.show()
```

```python
from itertools import combinations

combos = []
for r in range(1, 4):
    for combo in combinations(age_labels, r):
        combos.append(combo)

combo_labels = []
counts_m = []
counts_f = []

for combo in combos:
    mask = pd.Series([True] * len(df))
    for age in combo:
        mask = mask & df[timepoint_cols[age]].notna()
    for age in age_labels:
        if age not in combo:
            mask = mask & df[timepoint_cols[age]].isna()
    sub = df[mask]
    m = (sub['sex'] == 'M').sum()
    f = (sub['sex'] == 'F').sum()
    if m + f > 0:
        combo_labels.append(combo)
        counts_m.append(m)
        counts_f.append(f)

y_positions = np.arange(len(combo_labels))
dot_x = {age: i for i, age in enumerate(age_labels)}

fig, (ax_dot, ax_bar) = plt.subplots(1, 2, figsize=(8, 7),
                                      gridspec_kw={'width_ratios': [1, 3]})
for y, combo in enumerate(combo_labels):
    xs = [dot_x[a] for a in combo]
    ys = [y] * len(xs)
    ax_dot.plot(xs, ys, 'k-', linewidth=1.5, zorder=1)
    ax_dot.scatter(xs, ys, color='black', s=60, zorder=2)

ax_dot.set_xticks(range(len(age_labels)))
ax_dot.set_xticklabels(age_labels, rotation=90, fontsize=9)
ax_dot.set_yticks([])
ax_dot.set_ylim(-0.5, len(combo_labels) - 0.5)
ax_dot.spines[['top', 'right', 'left']].set_visible(False)

bar_h = 0.35
ax_bar.barh(y_positions + bar_h/2, counts_m, height=bar_h, color='black', label='M')
ax_bar.barh(y_positions - bar_h/2, counts_f, height=bar_h, color='gray', label='F')
ax_bar.set_yticks([])
ax_bar.set_xlabel('Sample size')
ax_bar.set_ylim(-0.5, len(combo_labels) - 0.5)
ax_bar.legend(frameon=False)
ax_bar.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```

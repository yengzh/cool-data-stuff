```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


samples = np.random.normal(loc=10, scale=2, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(samples, bins=30, density=True, color='steelblue', edgecolor='black', alpha=0.7, label='Samples')

x = np.linspace(samples.min(), samples.max(), 200)
plt.plot(x, norm.pdf(x, loc=10, scale=2), color='red', linewidth=2, label='Normal curve (μ=10, σ=2)')

plt.title('Normal Distribution (μ=10, σ=2)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
samples = np.random.uniform(low=5, high=15, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(samples, bins=30, density=True, color='steelblue', edgecolor='black', alpha=0.7, label='Samples')

plt.title(' uniform distribution ')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Dataset we're using
df = pd.read_csv("CFW.csv")
# 1 Months
khz = [4, 8, 12, 16, 24, 32]
onemonth = ['01mo_04_khz', '01mo_08_khz', '01mo_12_khz', '01mo_16_khz', '01mo_24_khz', '01mo_32_khz']

deaf_counts = []
not_deaf_counts = []

for col in onemonth:
    valid_data = df[col].dropna()
    deaf = (valid_data == 'NR').sum()
    deaf_counts.append(deaf)
    not_deaf = (valid_data != 'NR').sum()
    not_deaf_counts.append(not_deaf)
    
# imaging
plt.bar(khz, not_deaf_counts, color='blue', label='Not Deaf', width=2)
plt.bar(khz, deaf_counts, bottom=not_deaf_counts, color='red', label='Deaf (NR)', width=2)

plt.xlabel('Frequency (khz)')
plt.ylabel('number of Mice')
plt.title('tested @ 1 month')
plt.xticks(khz) # Force the x-axis to only show your tested frequencies

plt.show()
```

```python
df = pd.read_csv("CFW.csv")

cols =    ['01mo_04_khz', '01mo_08_khz', '01mo_12_khz', '01mo_16_khz', '01mo_24_khz', '01mo_32_khz']
colsone = ['06mo_04_khz', '06mo_08_khz', '06mo_12_khz', '06mo_16_khz', '06mo_24_khz', '06mo_32_khz']
colstwo = ['10mo_04_khz', '10mo_08_khz', '10mo_12_khz', '10mo_16_khz', '10mo_24_khz', '10mo_32_khz']

freqs = [4, 8, 12, 16, 24, 32]
ages = [('1 Month', cols, 'blue'), ('6 Month', colsone, 'orange'), ('10 Month', colstwo, 'red')]

fig, ax = plt.subplots(figsize=(7, 5))
for label, col_group, color in ages:
    data = df[col_group].replace('NR', float('nan')).apply(pd.to_numeric)
    means = data.mean()
    sems = data.sem()
    ax.errorbar(freqs, means, yerr=sems, label=label, color=color,
                marker='o', linewidth=2, capsize=4)

ax.set_xlabel('Frequency (kHz)')
ax.set_ylabel('dB SPL Threshold')
ax.set_xticks(freqs)
ax.set_ylim(0, 100)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the data and setup constants
df = pd.read_csv('CFW.csv')
n946 = pd.read_csv("946idk.csv")
khz = [4, 8, 12, 16, 24, 32]
x_positions = np.arange(len(khz)) 
width = 0.35 

onemonth = ['01mo_04_khz', '01mo_08_khz', '01mo_12_khz', '01mo_16_khz', '01mo_24_khz', '01mo_32_khz']
sixmonth = ['06mo_04_khz', '06mo_08_khz', '06mo_12_khz', '06mo_16_khz', '06mo_24_khz', '06mo_32_khz']
tenmonths = ['10mo_04_khz', '10mo_08_khz', '10mo_12_khz', '10mo_16_khz', '10mo_24_khz', '10mo_32_khz']

timepoints = {
    '1 month':onemonth,
    '6 months': sixmonth,
    '10 months': tenmonths,
}
datasets = {'cfw': df,'946':n946,}


# 3. Create the grid: 1 Row, 3 Columns. 
# figsize=(18, 5) makes the "poster board" wide enough to fit them all
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 20))

# 4. Loop through the timepoints, plotting each one on a different axis
for i, (time_title, columns) in enumerate(timepoints.items()):
    for j, (ds_name, data) in enumerate(datasets.items()):
        ax = axes[i][j]
        ax.set_title(f"{ds_name} - {time_title}")
        deaf_counts = []
        not_deaf_counts = []

    # Calculate counts just like before
        for col in columns:
            valid_data = data[col].dropna()
            deaf_counts.append((valid_data == 'NR').sum())
            not_deaf_counts.append((valid_data != 'NR').sum())
    
        ax.bar(khz, not_deaf_counts, color='blue', label='Not Deaf', width=2)
        ax.bar(khz, deaf_counts, bottom=not_deaf_counts, color='red', label='Deaf (NR)', width=2)
        ax.set_xlabel('Frequency (kHz)')
        ax.set_ylabel('Number of Mice')
        ax.set_xticks(khz)
        ax.legend()

#axes[2].legend()

plt.show()
```
